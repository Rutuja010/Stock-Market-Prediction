What is the moving average of the various stocks?

The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.

Advantages of Moving Averages:

Technical Analysis:Moving Averages help traders and investors understand the market's direction and make informed decisions based on price patterns.

Trend Confirmation: They confirm trends when prices consistently stay above or below certain moving averages.

Market Sentiment: Moving averages reflect overall market sentiment. A rising MA indicates bullish sentiment, while a falling MA indicates bearish sentiment.

```python
# Calculating the 20-day and 50-day moving averages directly on fdf
df['20 Day MA'] = df['Adj Close'].rolling(window=20).mean()
df['50 Day MA'] = df['Adj Close'].rolling(window=50).mean()

# Plot the Adjusted Closing Price and Moving Averages
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Adj Close'], label='Adjusted Close Price', color='blue')
plt.plot(df['Date'], df['20 Day MA'], label='20 Day Moving Average', color='red')
plt.plot(df['Date'], df['50 Day MA'], label='50 Day Moving Average', color='green')
plt.title('Infosys Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

What is a Price Distribution Histogram?
A price distribution histogram is a useful visualization to understand how frequently the stock's closing prices fall within certain ranges over a specified period.


```python
# Price distribution of closing price
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Interpreting the Histogram:
X-axis (Closing Price): This axis represents the different price ranges (bins). Each bin covers a range of prices (e.g., $300-$350, $350-$400).

Y-axis (Frequency): This axis shows the number of times the stock's closing price fell within each price range.

Bar Heights: The taller the bar, the more frequently the stock closed within that price range. For example, if the majority of bars are in the $350-$400 range, it suggests that the stock often closed within that range.

Here we are calculating the daily returns of the stock based on the adjusted closing prices.

fdf['Adj Close']: This is the column in the DataFrame fdf that contains the adjusted closing prices for each day.

<.pct_change()>: This function calculates the percentage change between the current adjusted closing price and the previous day adjusted closing price.
The result is stored in a new column 'Daily Return' in  fdf.

**Daily return**: Daily return is a measure that shows how much the price of a stock has changed from one day to the next, expressed as a percentage.

The daily return for a given day is calculated using the formula:

Daily Return = [(Price today - Price yesterday) / Price yesterday] * 100

Decompose the data for better visualization which contains trend , seasonality ,residual.




Trend Component: The overall direction of the data (long-term increase or decrease). Seasonal Component: The repeating short-term cycle (e.g., yearly, monthly). Residual Component: The remainder of the data after removing trend and seasonality (often considered noise).


```python
# Fill the missing values (if any )with forward_fill
df["Close"]=df["Close"].ffill()

# Ensure there is no NAN values in "Close" Column
df["Close"]=pd.to_numeric(df["Close"],errors="coerce")

# Drop th remaining null values
df=df.dropna(subset=["Close"])
```

```python
# Installing the required libraries
#!pip install statsmodels
from matplotlib import pyplot as plt

# Import the seasonal_decompose from statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform seasonal decompostion with an appropriate period for this dataset
# Assuming daily data with yearly seasonality (period = 252 for trading days , excluding weekends and holidays).
model = seasonal_decompose(df["Close"],model="multiplicative",period=252)

# plot the decompose components
Trend=model.trend
Seasonal=model.seasonal
Residual=model.resid

# Define the figure size

# plot the Trend
plt.figure(figsize=(8,6))
plt.plot(Trend,label="Trend",color="red")
plt.legend(loc="best") # Add legend with loc='best'to place the legend in a fixed location
plt.xticks(rotation=45)

# plot seasonal component
plt.figure(figsize=(8,6))
plt.plot(Seasonal,label="Seasonal",color="green")
plt.legend(loc='best')
plt.xticks(rotation=45)

# plot Residual component
plt.figure(figsize=(8,6))
plt.plot(Residual,label="Residual",color="blue")
plt.legend(loc='best')
plt.xticks(rotation=45)

```

```python
from arch import arch_model
import numpy as np
import matplotlib.pyplot as plt

# Rescaling factor
scaling_factor = 100

# Step 1: Inspecting and Rescaling the DataFrame
# Ensure that 'Daily_Return' has no missing values
df2 = df2.dropna(subset=['Daily_Return'])

# Rescale the 'Daily_Return' column
df2['Daily_Return_rescaled'] = df2['Daily_Return'] * scaling_factor

# Step 2: Fit the GARCH Model
# We are using GARCH(1,1) model for simplicity. You can try different values for (p,q) to optimize.
# p = 1 (for AR term), q = 1 (for MA term)
garch_model = arch_model(df2['Daily_Return_rescaled'], vol='Garch', p=1, q=1)

# Fit the model to the rescaled data
garch_result = garch_model.fit(disp='off')

# Step 3: Summary of the Model
# Display the result summary which provides insights on the model parameters
print(garch_result.summary())

# Step 4: Forecast Volatility
# Forecast volatility for the next 5 days (you can change the horizon as needed)
forecast_horizon = 5
garch_forecast = garch_result.forecast(horizon=forecast_horizon)

# Extract the variance forecasts (variance is volatility squared)
# variance.iloc[-1] gives the forecast for the most recent period
forecast_variance_rescaled = garch_forecast.variance.iloc[-1]

# Convert the forecasted variance to volatility (standard deviation) and then reverse the rescaling
forecast_volatility_rescaled = np.sqrt(forecast_variance_rescaled)
forecast_volatility = forecast_volatility_rescaled / scaling_factor

# Display forecasted volatility
print(f"Forecasted volatility for the next {forecast_horizon} days:\n", forecast_volatility)

# Step 5: Plotting Conditional Volatility
# Convert the conditional volatility back to the original scale for plotting
conditional_volatility_rescaled = garch_result.conditional_volatility
conditional_volatility = conditional_volatility_rescaled / scaling_factor

plt.figure(figsize=(10, 6))
plt.plot(df2['Date'], conditional_volatility, label='Conditional Volatility')
plt.title('GARCH Model - Conditional Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
import streamlit as st
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.linear_model import LinearRegression
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from arch import arch_model
from prophet import Prophet

# Load stock data
st.title("Stock Price Prediction using Multiple Models")
ticker = st.text_input("Enter Stock Ticker", "AAPL")
data_load_state = st.text("Loading data...")

df = yf.download(ticker, period='5y')
df['Date'] = df.index
df = df[['Date', 'Adj Close']]
data_load_state.text("Data Loaded!")

st.subheader(f"{ticker} Stock Price Data")
st.write(df.tail())

# Select model for prediction
model_option = st.selectbox(
    "Choose a prediction model",
    ("ARIMA", "SARIMA", "LSTM", "Linear Regression", "GARCH", "Prophet")
)

forecast_days = st.slider("Forecast days:", 1, 30, 10)

# ARIMA Model
if model_option == "ARIMA":
    st.subheader("ARIMA Model Forecast")
    model_arima = ARIMA(df['Adj Close'], order=(1, 1, 1))
    model_arima_fit = model_arima.fit()
    st.text(model_arima_fit.summary())

    forecast_arima = model_arima_fit.forecast(steps=forecast_days)

    # Plot ARIMA results
    plt.figure(figsize=(10, 5))
    plt.plot(df['Date'], df['Adj Close'], label='Actual Prices')
    forecast_dates = pd.date_range(df['Date'].iloc[-1], periods=forecast_days+1)[1:]
    plt.plot(forecast_dates, forecast_arima, label='Forecasted Prices', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    st.pyplot(plt)

# SARIMA Model
elif model_option == "SARIMA":
    st.subheader("SARIMA Model Forecast")
    model_sarima = SARIMAX(df['Adj Close'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_sarima_fit = model_sarima.fit()
    st.text(model_sarima_fit.summary())

    forecast_sarima = model_sarima_fit.forecast(steps=forecast_days)

    # Plot SARIMA results
    plt.figure(figsize=(10, 5))
    plt.plot(df['Date'], df['Adj Close'], label='Actual Prices')
    forecast_dates = pd.date_range(df['Date'].iloc[-1], periods=forecast_days+1)[1:]
    plt.plot(forecast_dates, forecast_sarima, label='Forecasted Prices', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    st.pyplot(plt)

# LSTM Model
elif model_option == "LSTM":
    st.subheader("LSTM Model Forecast")

    # Data preprocessing
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df['Adj Close'].values.reshape(-1, 1))

    X_train, y_train = [], []
    for i in range(60, len(scaled_data)):
        X_train.append(scaled_data[i-60:i, 0])
        y_train.append(scaled_data[i, 0])

    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Build the LSTM model
    model_lstm = Sequential()
    model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model_lstm.add(Dropout(0.2))
    model_lstm.add(LSTM(units=50, return_sequences=False))
    model_lstm.add(Dropout(0.2))
    model_lstm.add(Dense(units=1))

    model_lstm.compile(optimizer='adam', loss='mean_squared_error')
    model_lstm.fit(X_train, y_train, epochs=5, batch_size=32)

    predicted_lstm = model_lstm.predict(X_train[-forecast_days:])
    predicted_lstm = scaler.inverse_transform(predicted_lstm)

    # Plot LSTM results
    plt.figure(figsize=(10, 5))
    plt.plot(df['Date'], df['Adj Close'], label='Actual Prices')
    future_dates = pd.date_range(df['Date'].iloc[-1], periods=forecast_days+1)[1:]
    plt.plot(future_dates, predicted_lstm, label='Forecasted Prices', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    st.pyplot(plt)

# Linear Regression Model
elif model_option == "Linear Regression":
    st.subheader("Linear Regression Model Forecast")

    df['Date_ordinal'] = pd.to_datetime(df['Date']).apply(lambda date: date.toordinal())
    X = df['Date_ordinal'].values.reshape(-1, 1)
    y = df['Adj Close'].values

    model_lr = LinearRegression()
    model_lr.fit(X, y)

    future_dates = pd.date_range(df['Date'].iloc[-1], periods=forecast_days+1).to_pydatetime()
    future_dates_ordinal = np.array([date.toordinal() for date in future_dates]).reshape(-1, 1)

    predicted_lr = model_lr.predict(future_dates_ordinal)

    # Plot Linear Regression results
    plt.figure(figsize=(10, 5))
    plt.plot(df['Date'], df['Adj Close'], label='Actual Prices')
    plt.plot(future_dates, predicted_lr, label='Forecasted Prices', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    st.pyplot(plt)

# GARCH Model
elif model_option == "GARCH":
    st.subheader("GARCH Model Forecast")

    returns = 100 * df['Adj Close'].pct_change().dropna()

    model_garch = arch_model(returns, vol='Garch', p=1, q=1)
    model_garch_fit = model_garch.fit(disp='off')

    forecast_garch = model_garch_fit.forecast(horizon=forecast_days)

    # Plot GARCH results
    plt.figure(figsize=(10, 5))
    plt.plot(forecast_garch.variance.iloc[-1], label='Forecasted Volatility', color='red')
    plt.xlabel('Date')
    plt.ylabel('Variance')
    plt.legend()
    st.pyplot(plt)
```
